```python
print("Metode Konvensional:")
for x in range(-4, 5):
    print(f"f({x}) = {3*x - 7}")
```

```python
def f(x):
    return 3*x - 7
print("Menggunakan def:")
for x in range(-4, 5):
    print(f"f({x}) = {f(x)}")
```

```python
def fx (x):
  return (3 * x) - 7
domain = [i for i in range (-4,5)]
for j in domain:
  print ("f(",j,") = ",fx(j))
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-100, 100, 400)
y = 2 * x

plt.figure(figsize=(8,5))
plt.plot(x, y, label="f(x) = 2x", color="blue")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title("Grafik f(x) = 2x", color="blue")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-19.9, 19.9, 400)
y = x**2 + 3*x - 10

plt.figure(figsize=(8,5))
plt.plot(x, y, label="f(x) = x² + 3x - 10", color="green")
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title("Grafik f(x) = x² + 3x - 10", color="green")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
